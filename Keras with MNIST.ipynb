```python
import tensorflow as tf
mnist=tf.keras.datasets.mnist
```

```python
(train_x,train_y),(test_x,test_y)=mnist.load_data()
```

```python
train_x,test_x=train_x/255.0,test_x/255.0
```

```python
model=tf.keras.models.Sequential([tf.keras.layers.Flatten(input_shape=(28,28)),
                                 tf.keras.layers.Dense(512,activation=tf.nn.relu),
                                 tf.keras.layers.Dropout(0.2),
                                 tf.keras.layers.Dense(10,activation=tf.nn.softmax)])
```

```python
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(train_x,train_y,epochs=5)
model.evaluate(test_x,test_y)
```

To avoid overfitting..techniques such as regularization-dropout. Only while training, we deactivate the part of the neurons and while testing reactivate them back. This is to reduce the overfitting problem. This is mainly performed when training accuracy is high and testing accuracy is low.

## MNIST Multilayer Variant using Keras

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Activation,Dense,Dropout,Flatten
mnist=tf.keras.datasets.mnist
```

```python
(train_x,train_y),(test_x,test_y)=mnist.load_data()
train_x,test_x=train_x/255.0,test_x/255.0
```

```python
train_x.shape,train_y.shape,test_x.shape,test_y.shape
```

#### create a sequential model and add layers

```python
model=Sequential()
```

#### add layers (play around with the layers to change the acc)

```python
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(200,activation='relu'))
model.add(Dense(100,activation='relu'))
model.add(Dense(50,activation='relu'))
model.add(Dense(30,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(train_x,train_y,epochs=5)
model.evaluate(test_x,test_y)
```
